### "Maternal Mortality in the U.S.: How do we compare to the rest of the world?"

```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing datasets

"""
https://data.unicef.org/resources/data_explorer/unicef_f/?ag=UNICEF&df=GLOBAL_DATAFLOW&ver=1.0&dq=.MNCH_MATERNAL_DEATHS+MNCH_LTR_MATERNAL_DEATH+MNCH_MMR+COD_MATERNAL_CONDITIONS+MNCH_ANC1+MNCH_ANC4+MNCH_SAB.F.&startPeriod=1970&endPeriod=2022
UNICEF Global Maternal Mortality rates
"""
global_mm_rate = pd.read_excel("GLOBAL_DATAFLOW_1970-2022.xlsx")

```

### A deeper look by Country
- Which countries are the top/bottom 5 for maternal deaths?
- How does the U.S. compare to those top/bottom 5?
- How has the U.S. individually done over time?

```python
"""
Taking a look at maternal mortality rates for each country
summed over the past 5 years.

Using UNICEF data"""

years_mm_rate = [str(year) for year in range(2013,2018)]

filtered_data = global_mm_rate[(global_mm_rate["TIME_PERIOD"].isin(years))\
                               &(global_mm_rate["Indicator"] ==\
                                 "Maternal mortality ratio (number of maternal deaths per 100,000 live births)")]
past_five_years_rates = filtered_data.groupby("Geographic area")["OBS_VALUE"].mean().sort_values() #returning the avg mortality rate for time period
```

```python
#Taking a look at the TOP 5 COUNTRIES with LOWEST Maternal death rates
lowest_rate = filtered_data.groupby("Geographic area")["OBS_VALUE"].mean().sort_values().head(5)
```

```python
lowest_rate
```

```python
#Taking a look at the TOP 5 COUNTRIES with HIGHEST Maternal death rates
highest_rate = filtered_data.groupby("Geographic area")["OBS_VALUE"].mean().sort_values().tail(5)
```

```python
highest_rate
```

```python
## U.S. Maternal Death rate for the past 5 years
us = filtered_data.groupby("Geographic area")["OBS_VALUE"].mean().loc["United States"]
```

```python
us
```

```python
#concatenate the highest and lowest to plot them together
high_low_rates = pd.concat([lowest_rate, highest_rate])
```

```python
high_low_rates
```

```python
high_low_rates.plot(kind="barh", figsize = (12,10))
```

```python
#chart comparing the US to the top 5 countries with lowest rates
lowest_rate.loc["United States"] = us
```

```python
lowest_rate.plot(kind="barh", title="Top 5 Countries with Lowest Maternal-related deaths vs. the U.S.", ylabel="# of Materna-related Deaths", xlabel="Country")
```

```python
#chart comparing the US to the top 5 countries with highest rates
highest_rate.loc["United States"] = us
```

```python
highest_rate.plot(kind="barh", title="Top 5 Countries with Highest Maternal-related deaths vs. the U.S.", xlabel="Country")
```

```python
"""
Now I want to look at the countries closest to the U.S.
in Maternal death rates by +/- 5
"""

past_fiveyears_rates_plusminusfive = past_five_years_rates[(past_five_years_rates < (us + 5)) & (past_five_years_rates > (us - 5))]
```

```python
past_fiveyears_rates_plusminusfive
```

```python
#dropping North America as it's a region not a country
past_fiveyears_rates_plusminusfive.drop("North America", inplace=True)
```

```python
past_fiveyears_rates_plusminusfive.plot(kind="barh", title="Countries closest in MMR to U.S. from 2013-2018", xlabel="Country")
```

### Skilled professional in attendance of birth?

```python
"""
Using the UNICEF data

Indicator = Skilled birth attendant - percentage of deliveries attended by skilled health personnel"""
```

```python
birth_attendance = global_mm_rate[\
  (global_mm_rate["Indicator"] == \
               "Skilled birth attendant - percentage of deliveries attended by skilled health personnel")\
   &  (global_mm_rate["TIME_PERIOD"].isin([str(year) for year in range(2013,2018)]))]\
.groupby(["Geographic area"])["OBS_VALUE"].mean()
```

```python
birth_attendance
```

```python
"""
looking at attendance of skilled professional at births
against mortality ratio by geographic area

"""
mortality = global_mm_rate[\
  (global_mm_rate["Indicator"] == \
               "Maternal mortality ratio (number of maternal deaths per 100,000 live births)")\
   &  (global_mm_rate["TIME_PERIOD"].isin([str(year) for year in range(2013,2018)]))]\
.groupby(["Geographic area"])["OBS_VALUE"].mean()
```

```python
mortality
```

```python
attendance_and_mortality = pd.merge(mortality.reset_index(), birth_attendance.reset_index(), on="Geographic area", suffixes=["_MORT", "_ATTEND"])
#obs_valuex= mortality obs_valuey=birthattendance
```

```python
attendance_and_mortality
```

```python
attendance_and_mortality.plot(kind="scatter", x="OBS_VALUE_MORT", y="OBS_VALUE_ATTEND", figsize=(14,12))
```

```python
#bar chart comparing top and bottom 5 countries against skilled birth attendance rate
```
